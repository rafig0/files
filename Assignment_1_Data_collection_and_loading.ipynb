#Load data from CSV, JSON, Excel, or online APIs into Python using pandas.Explore dataset dimensions, data types, and summary.

```python
import pandas as pd
```

# ================================
# Loading Data with Pandas
# ================================

# 1. Load Data from a CSV file

```python

print("---- Loading CSV File ----")
csv_data = pd.read_csv("sample_data.csv")   # Replace with your file path
print(csv_data.head(), "\n")
```

# 2. Load Data from a JSON file

```python

print("---- Loading JSON File ----")
json_data = pd.read_json("sample_data.json")  # Replace with your file path
print(json_data.head(), "\n")
```

# 3. Load Data from an Excel file

```python

print("---- Loading Excel File ----")
excel_data = pd.read_excel("sample_data.xlsx", sheet_name="Sheet1")  # Replace with your file path
print(excel_data.head(), "\n")
```

# 4. Load Data from an Online API (using pandas.read_json or pd.DataFrame directly)
## Example: Public API for random users

```python

print("---- Loading Data from Online API ----")
url = "https://jsonplaceholder.typicode.com/users"
api_data = pd.read_json(url)
print(api_data.head(), "\n")
```

# ================================
# Exploring the Dataset

```python
# Shape (rows, columns)
print("CSV Data Shape:", csv_data.shape)
```

```python
# Column names
print("CSV Columns:", csv_data.columns.tolist())
```

```python
# Summary statistics (numerical columns)
print("\nSummary Statistics:\n", csv_data.describe())
```
