```python
plt.figure(figsize=(8,6))
plt.hist(df['Glucose'], bins=15, color='skyblue', edgecolor='black')
plt.title("Histogram of Glucose", fontsize=16)
plt.xlabel("Glucose Level")
plt.ylabel("Frequency")
plt.show()
```

**What are bins in a histogram?**

A bin represents an interval (range) of values in the dataset.

When plotting a histogram, data values are grouped into these bins.

Each bar in the histogram shows how many data points fall into that bin.

**Example:**

Suppose Glucose values range from 0 to 200.

If bins=10, the range will be divided into 10 equal intervals:

0–20, 20–40, 40–60, …, 180–200.

The histogram will show how many people have glucose levels in each interval.

```python
# Plot histograms for all numeric columns
df.hist(bins=15, figsize=(15,12), color='skyblue', edgecolor='black')
plt.suptitle("Histograms of Diabetes Dataset Features", fontsize=18)
plt.show()
#Pandas automatically finds all numeric columns in your dataset and draws histograms for each.
```

#Boxplot
Why use a Boxplot?

A boxplot is used to visualize the distribution of a numeric variable and detect outliers.

```python
plt.figure(figsize=(8,6))
sns.boxplot(y=df['Glucose'], color='lightgreen')
plt.title("Boxplot of Glucose", fontsize=16)
plt.ylabel("Glucose Level")
plt.show()
```

```python
# What is palette in Seaborn?
# In Seaborn, palette controls the colors of your plot elements (bars, boxes, points, etc.).
# Example:
# sns.boxplot(data=df, palette="Pastel1")

# "Pastel1" → a predefined color palette with soft pastel colors.
# Other options:
# "Set2", "coolwarm", "viridis", "Blues", "rainbow", etc.


plt.figure(figsize=(15,10))
sns.boxplot(data=df, palette="Pastel1")
plt.title("Boxplots of All Features", fontsize=18)
plt.xticks(rotation=45)  # Rotate x-axis labels for readability
plt.show()
```

#Scatterplot


What is hue in Seaborn?

In Seaborn plots, hue is used to separate data by a categorical variable and give each category a different color.

Example
sns.boxplot(x="Age", y="BMI", hue="Outcome", data=df, palette="Pastel1")


x → horizontal axis (Age)

y → vertical axis (BMI)

hue="Outcome" → divides the data into groups based on Outcome column (e.g., 0 = No Diabetes, 1 = Diabetes)

Seaborn automatically assigns different colors for each hue category.

```python
plt.figure(figsize=(8,6))
sns.scatterplot(x='Glucose', y='BMI', hue='Outcome', data=df, palette='coolwarm')
plt.title("Scatterplot: Glucose vs BMI by Outcome", fontsize=16)
plt.xlabel("Glucose")
plt.ylabel("BMI")
plt.show()
```

```python
corr_matrix = df.corr()
```

**1️⃣ annot=True**

annot stands for “annotations”.

When True, the numerical values of the correlation matrix (or data) are displayed inside each cell of the heatmap.

If False, only colored squares are shown.

**2️⃣ cmap='YlGnBu'**

cmap = color map → defines the colors used in the heatmap.

'YlGnBu' = Yellow → Green → Blue gradient.

You can choose other color maps too, e.g.:

'coolwarm', 'viridis', 'Blues', 'RdYlBu'

Purpose: visually show low vs high values in different colors.

**3️⃣ fmt=".2f"**

fmt = format of the annotated numbers.

".2f" → show floating point numbers rounded to 2 decimal places.

Example: 0.34567 → displayed as 0.35

```python
plt.figure(figsize=(10,8))
sns.heatmap(corr_matrix, annot=True, cmap='YlGnBu', fmt=".2f")
plt.title("Correlation Heatmap of Features", fontsize=16)
plt.show()
```

#Pairplot

```python
sns.pairplot(df, hue='Outcome', diag_kind='hist', palette='coolwarm')
plt.suptitle("Pairplot of Features by Outcome", y=1.02, fontsize=16)
plt.show()
```

1️⃣ What is diag_kind?

diag_kind controls the type of plot used on the diagonal of the pairplot.

The diagonal shows the distribution of a single variable.

2️⃣ Options for diag_kind

'hist' → Use histograms on the diagonal.

Shows frequency/count of values for each feature.

'kde' → Use Kernel Density Estimate curves on the diagonal.

Smooth curve showing distribution (good for continuous data).
