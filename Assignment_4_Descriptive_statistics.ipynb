#Correlation
###Correlation is a concept — it tells us how two variables are related.

Correlation coefficient is the numerical value that quantifies that correlation.

So when you “calculate correlation” in Python, what the functions (np.corrcoef or pandas.Series.corr) actually give you is the correlation coefficient.

In other words:

Saying “correlation” without a number is theoretical.

To use it practically (and show in Python), you always calculate the correlation coefficient, which tells you:

+1 → perfect positive correlation

-1 → perfect negative correlation

0 → no correlation

So when you see a single number from df['data1'].corr(df['data2']) or np.corrcoef(data1, data2)[0,1], that is the correlation — Python just expresses it numerically as the coefficient.

```python
import pandas as pd
df = pd.DataFrame({'data1': [10, 20, 30, 40, 50],
                   'data2': [15, 25, 35, 45, 55]})
correlation = df['data1'].corr(df['data2'])
print('Correlation coefficient:', correlation)

```

#Covariance


Exactly! Covariance works the same way conceptually:

Covariance tells us how two variables vary together.

But the output from Python (np.cov) is a covariance matrix, which shows covariance between all pairs of variables.

If you only want the covariance between two variables, you pick the off-diagonal element of the matrix.

Example:

```python
import numpy as np

data1 = [10, 20, 30, 40, 50]
data2 = [15, 25, 35, 45, 55]

cov_matrix = np.cov(data1, data2)
covariance = cov_matrix[0, 1]  # covariance between data1 and data2
print('Covariance:', covariance)
```
