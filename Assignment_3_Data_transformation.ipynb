#What is Normalization?

###Normalization is a data preprocessing technique used to rescale numerical features so they fall within a specific range, typically [0, 1] or [-1, 1].

It helps ensure that no single feature dominates others due to large magnitude differences (e.g., salary vs. age).

Especially useful for distance-based algorithms such as:

K-Nearest Neighbors (KNN)

Neural Networks

Gradient Descentbased models

Why Normalization is Needed

Different columns in a dataset may have different units and ranges.
Example:

Age = 2060

Salary = 20,000100,000

If unnormalized, the larger-valued feature (Salary) can dominate model training.

Normalization equalizes the scale of all features, improving accuracy and training stability.

```python
# Simple Example: One Hot Encoding using sklearn

import pandas as pd
from sklearn.preprocessing import OneHotEncoder

# Sample dataset
data = {'City': ['Pune', 'Mumbai', 'Delhi', 'Pune']}
df = pd.DataFrame(data)
print("Original Data:")
print(df)

# Initialize the encoder
encoder = OneHotEncoder(sparse_output=False) #scikit-learn to return a dense NumPy array instead of a sparse matrix.

# Transform the data
encoded = encoder.fit_transform(df[['City']])

# Convert encoded data to DataFrame
encoded_df = pd.DataFrame(encoded, columns=encoder.get_feature_names_out(['City']))

# Combine with original data
final_df = pd.concat([df, encoded_df], axis=1)

print("\nOne Hot Encoded Data:")
print(final_df)
```

```python
import pandas as pd

# Sample Data
data = {'City': ['Pune', 'Mumbai', 'Delhi', 'Pune']}
df = pd.DataFrame(data)

# One Hot Encoding in Binary Format
encoded_df = pd.get_dummies(df, columns=['City']).astype(int)

print(encoded_df)
```
