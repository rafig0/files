#Handle missing data (mean, median, mode),remove duplicates, and fix data types. Identify and treat outliers.

# ================================
# Data Cleaning with Pandas
# ================================

```python
import pandas as pd
import numpy as np

# Sample dataset with issues
data = {
    "Student_ID": [1, 2, 3, 4, 5, 5],   # Duplicate ID (5)
    "Name": ["Amit", "Riya", "Sahil", "Neha", "Ravi", "Ravi"],
    "Age": [20, np.nan, 22, 21, 200, 21],   # Missing + Outlier (200 years!)
    "Marks": [85, 90, np.nan, 88, 95, 95],  # Missing values
    "Joining_Date": ["2023-06-01", "2023-06-05", "06/10/2023", "2023/06/15", "15-06-2023", "2023-06-15"]
}
```

```python
df = pd.DataFrame(data)
print("---- Original Data ----")
print(df, "\n")
```

# ================================
# 1. Handling Missing Data
# ================================

```python
print("---- Handling Missing Data ----")

# Fill Age with mean
df["Age"].fillna(df["Age"].mean(), inplace=True)
print(df, "\n")
```

# ================================
# 2. Remove Duplicates
# ================================



```python
print("---- Removing Duplicates ----")
df = df.drop_duplicates()
print(df, "\n")
```

# ================================
# 3. Fix Data Types
# ================================

```python

print("---- Fixing Data Types ----")

# Convert Joining_Date to datetime
df["Joining_Date"] = pd.to_datetime(df["Joining_Date"], dayfirst=True, errors="coerce")
```

```python
# Ensure Age is integer
df["Age"] = df["Age"].astype(int)

print(df.dtypes, "\n")
print(df, "\n")
```

# ================================
# 4. Identify and Treat Outliers
# ================================

```python


print("---- Handling Outliers ----")

# Using IQR (Interquartile Range) method
Q1 = df["Age"].quantile(0.25)
Q3 = df["Age"].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

print("Lower Bound:", lower_bound, "Upper Bound:", upper_bound)

# Treat outliers: replace with median
df.loc[df["Age"] > upper_bound, "Age"] = df["Age"].median()

print("\nFinal Cleaned Dataset:")
print(df)
```
